```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
analysis_df = pd.read_pickle('TRex_analysis_pd.pkl')
```

```python
analysis_df
```

```python
vel_df = analysis_df['velocity'].apply(pd.Series).transpose()
```

```python
fig = plt.figure()
time = np.linspace(0,30,1799)
plt.plot(time, vel_df['ILR2K'], linewidth = 0.1)
plt.plot(time, vel_df['ILR2K'].median(axis = 1),'k', label='oxtr KO')

plt.xlabel('time [minutes]', fontsize=18)
plt.ylabel('velocity [pixels/second]', fontsize=16)
plt.ylim(0, 800)

fig = plt.figure()

plt.plot(time, vel_df['ILR2w'], linewidth = 0.1)
plt.plot(time, vel_df['ILR2w'].median(axis = 1),'k', label='oxtr wt')

plt.xlabel('time [minutes]', fontsize=18)
plt.ylabel('velocity [pixels/second]', fontsize=16)
plt.ylim(0, 800)
plt.legend()
```

```python
fig = plt.figure()
time = np.linspace(0,1799,1799)
plt.plot(time, vel_df['ILR1K'].median(axis = 1),'m', label='oxtrl KO')
plt.fill_between(time, vel_df['ILR1K'].median(axis = 1) - vel_df['ILR1K'].std(axis = 1),
         vel_df['ILR1K'].median(axis = 1) + vel_df['ILR1K'].std(axis = 1), fc = 'm')

plt.plot(vel_df['ILR1w'].median(axis = 1), 'g', label = 'oxtrl WT')
plt.fill_between(time, vel_df['ILR1w'].median(axis = 1) - vel_df['ILR1w'].std(axis = 1),
         vel_df['ILR1w'].median(axis = 1) + vel_df['ILR1w'].std(axis = 1), fc = 'g')
fig.suptitle('velocity over time', fontsize=20)
plt.xlabel('time [seconds]', fontsize=18)
plt.ylabel('velocity [pixels/second]', fontsize=16)
plt.ylim(0, 600)
plt.legend()
#fig.savefig('velocity_ILR1.png')
```

```python
fig = plt.figure()
time = np.linspace(0,1799,1799)
plt.plot(time, vel_df['ILR2K'].median(axis = 1),'m', label='ILR2 KO')
plt.fill_between(time, vel_df['ILR2K'].median(axis = 1) - vel_df['ILR2K'].std(axis = 1),
         vel_df['ILR2K'].median(axis = 1) + vel_df['ILR2K'].std(axis = 1), fc = 'm')

plt.plot(vel_df['ILR2w'].median(axis = 1), 'g', label = 'ILR2 WT')
plt.fill_between(time, vel_df['ILR2w'].median(axis = 1) - vel_df['ILR2w'].std(axis = 1),
         vel_df['ILR2w'].median(axis = 1) + vel_df['ILR2w'].std(axis = 1), fc = 'g')
fig.suptitle('velocity over time', fontsize=20)
plt.xlabel('time [seconds]', fontsize=18)
plt.ylabel('velocity [pixels/second]', fontsize=16)
plt.ylim(0, 600)
plt.legend()
#fig.savefig('velocity_ILR2.png')
```

```python
dist_df = analysis_df['neighbor'].apply(pd.Series).transpose()
```
